```python
#General libraries

import numpy as np
import pandas as pd

#Visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "D:\programming project feb 2024\Spain Energy Dataset\df_viz.csv"
#Reading our Dataset
df_ml=pd.read_csv(path,index_col='time',)
#Showing first five values from the dataset
df_ml.head()
```

## Train Test splitting

```python
# creating train test data sets
from sklearn.model_selection import train_test_split
x = df_ml.drop(['price_actual','year','month','month_name','weekdays'], axis=1)
y = df_ml['price_actual']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)
```

```python
# Model Creation
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, mean_absolute_error, r2_score
```

### Linear Regression

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
lr_predictions = lr_model.predict(x_test)
print("Linear Regression")
print("------------------------------------")
#mae = mean_absolute_error(y_train, lr_model.predict(x_train))
#print("MAE training:", mae)
#print("------------------------------------")
lr_mae = mean_absolute_error(y_test, lr_predictions)
print("MAE:", lr_mae)
print("------------------------------------")
lr_r2 = r2_score(y_test, lr_predictions)
print("R-squared:", lr_r2)


```

### Ridge

```python
r_model = Ridge()
r_model.fit(x_train, y_train)
r_predictions = r_model.predict(x_test)
print("Ridge")
print("------------------------------------")
r_mae = mean_absolute_error(y_test, r_predictions)
print("MAE:", r_mae)
print("------------------------------------")
r_r2 = r2_score(y_test, r_predictions)
print("R-squared:", r_r2)
```

### Random Forest Regressor

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(x_train, y_train)
rf_predictions = rf_model.predict(x_test)
print("Random Forest ")
print("------------------------------------")
rf_mae = mean_absolute_error(y_test, rf_predictions)
print("MAE:", rf_mae)
print("------------------------------------")
rf_r2 = r2_score(y_test, rf_predictions)
print("R-squared:", rf_r2)
```
