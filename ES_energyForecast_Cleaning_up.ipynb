```python
#General Libraries
import pandas as pd
import numpy as np
#Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'D:\programming project feb 2024\Spain Energy Dataset\energy_dataset.csv'
#Reading our Dataset
df_electricity=pd.read_csv(path)
#Showing first five values from the dataset
df_electricity.head()
```

```python
#Checking the shape of our dataset
print(df_electricity.shape)
#Checking the type of data and how
print(df_electricity.info())
```

As we can see, columns 10 (generation hydro pumped storage aggregated) and 23 (forecast wind offshore day ahead) have all null values. So we can drop those columns.

```python
df_electricity.isna().sum()
```

```python
df_electricity.describe().T
```

We can see above that we have some more columns with 0 values for all the records.So, we will drop 8 columns in total from our dataset.

```python

df_electricity_cleaned=df_electricity.copy()
df_electricity_cleaned.columns=df_electricity_cleaned.columns.str.replace(' ','_').str.replace('/','_')
```

```python
# Drop unusable columns
df_electricity_cleaned = df_electricity_cleaned.drop(['generation_fossil_coal-derived_gas',
                                                'generation_fossil_oil_shale',
                                                'generation_fossil_peat',
                                                'generation_geothermal',
                                                'generation_hydro_pumped_storage_aggregated',
                                                'generation_marine',
                                                'generation_wind_offshore',
                                                'forecast_wind_offshore_eday_ahead',
                                                'forecast_solar_day_ahead',
                                                'forecast_wind_onshore_day_ahead',
                                                'total_load_forecast'], axis=1)
```

```python
df_electricity_cleaned.head()
```

```python
df_electricity_cleaned.set_index('time', inplace=True)
df_electricity_cleaned.head()
```

```python
df_electricity_cleaned.isna().sum()
```

As, our percentage of missing values in each column is very low.So, I am going to fillna value with the method of forward fill.The ffill is a good choice because our data is not increasing linearly/polynomially.In most of the feature is the generation

```python
df_electricity_cleaned.fillna(method='ffill',inplace=True)
df_electricity_cleaned.head()
```

```python
df_electricity_cleaned.isnull().sum()
```

```python
df_electricity_cleaned.to_csv('D:\programming project feb 2024\Spain Energy Dataset\df_electricity_cleaned.csv')
```
